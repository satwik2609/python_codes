```python
x = 5
def change():
    x = x**2   #we can't change global variable
    print(x)
change()
```

```python
x = 5
def change():
    p = x**2   #we can access global variable
    print(p)
change()
```

```python
x = 5
def change():
    global x  #used to change global variable
    x = x**2   
print("Before function: ",x)
change()
print("After function: ", x)
```

```python
def hello():
    print("Hello world")
    hello()
hello()
#RecursionError--> cz hello() function keeps making memory spaces for every calling and runs untill memory is filled completely
```

```python
def hello():
    print("Hello world")
    hello()
hello()
while True:
    hello()
#hello() function keeps clearing memory space after every calling so it doesn't stop and keeps running infinite times
```
